# Iris example

Load Iris dataset:

```python
from sklearn import datasets
iris = datasets.load_iris()
```

Show data of Iris dataset:

```python
import pandas as pd
df_iris = pd.DataFrame(iris["data"],columns=iris["feature_names"])
df_iris["species_number"] = iris["target"]
species_map = dict(zip(range(3),iris["target_names"]))
df_iris["species"] = [species_map[x] for x in iris["target"]]
df_iris
```

Scatterplot data for different feature sets:

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="ticks")

sns.pairplot(df_iris.drop("species_number",axis=1), hue="species")
```
